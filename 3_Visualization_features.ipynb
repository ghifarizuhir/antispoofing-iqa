```python
# props : untuk membuat titik rata-rata pada boxplot

props = {'marker': 'o',
         'markeredgecolor': 'black',
         'markerfacecolor': 'white',
         'markersize': 7}
```

```python
fitur = ['mse00', 'psnr01', 'ad02', 'sc03', 'nk04', 'md05', 'lmse06', 'nae07',
        'snrv08', 'ramdv09', 'mas10', 'mams11', 'sme12', 'gme16', 'gpe17',
        'ssim18', 'vif19', 'hlfi25']
```

```python
df = pd.read_csv("/content/drive/dataset.csv")
```

```python
df.columns
```

```python
class_map = {"real" : "real",
           "paper" : "attack",
           "replay" : "attack"}

df['class'].replace(class_map, inplace=True)
```

```python
dev_map = {1 : "Samsung",
           2 : "HTC Desire EYE",
           3 : "MEIZU X5",
           4 : "ASUS",
           5 : "Sony",
           6 : "Oppo N3"}

df['device'].replace(dev_map, inplace=True)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=6, figsize=(100,60))

for i in range(3):
  for j in range(6):
    idx = (i*6)+j
    sns.boxplot(data =df, x=fitur[idx], y="class",
                palette="Set2", showmeans=True, meanprops= props,
                ax=axes[i, j])
```

```python
sns.catplot(data=df, x="class", y="lmse06", col="device",
            palette="Set2", kind = 'box')
sns.set(font_scale=2)
```
